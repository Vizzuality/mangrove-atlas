```python
import pandas as pd
import numpy as np
import geopandas as gpd
import fiona
import folium
```

## 1) Download data if needed  
Data is stored in the GC bucket as Geojson.

```python
data_url = 'https://storage.googleapis.com/mangrove_atlas/widget_data'
vector_file = 'alert_region_tiles.geojson'
data = gpd.read_file(f'{data_url}/{vector_file}')
data
```

```python
map = folium.Map(location=[0,0], tiles="OpenStreetMap", zoom_start=3)
folium.GeoJson(data).add_to(map)
map
```

```python
out_path = '../../../../data/alert_region_tiles.geojson'

data.to_file(out_path, driver="GeoJSON") 
```
